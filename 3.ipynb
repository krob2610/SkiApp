```python
#Round time to decisecond in time column and drop row if it already has the same timestamp (time is giving in pd.datetime)

def unify_dates(df: pd.DataFrame) -> pd.DataFrame:
    df["time"] = pd.to_datetime(df["time"], format='%Y-%m-%d %H:%M:%S.%f')  # Convert to datetime
    df["time"] = df["time"].dt.round('100L')  # Round to nearest decisecond
    df.drop_duplicates(subset="time", inplace=True)
    return df
```
